```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/chatgpt_reviews (1).csv')
```

```python
df['Review Date'] = pd.to_datetime(df['Review Date'])
df.info()
```

Here We Replaced Null values in Review with the text 'No Review' so that we dont have any missing values and perform a better analysis

```python
df['Review'] = df['Review'].fillna('No Review')
```

```python


plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Ratings', data=df, hue='Ratings', palette='viridis', legend=False)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')

# Add annotations
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
df['Year'] = df['Review Date'].dt.year
year_wise_ratings = df.groupby('Year')['Ratings'].count().reset_index()
year_wise_ratings.rename(columns={'Ratings': 'Number of Reviews'}, inplace=True)
```

```python
plt.figure(figsize=(10, 8))
ax = sns.barplot(x='Year', y='Number of Reviews', data=year_wise_ratings, hue='Year', palette='coolwarm', legend=False) # Changed to barplot as requested
plt.title('Number of Reviews per Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add annotations for bar plot
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
year_rating_counts = df.groupby(['Year', 'Ratings']).size().reset_index(name='Count')

plt.figure(figsize=(12, 7))
ax = sns.barplot(x='Year', y='Count', hue='Ratings', data=year_rating_counts, palette='viridis')
plt.title('Rating-wise Comparison per Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Rating')

# Add annotations
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')
```

```python
# Initialize the VADER sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

def get_sentiment(text):
    if not isinstance(text, str): # Handle non-string types, e.g., if there are still non-string values
        return 'neutral'
    score = sia.polarity_scores(text)
    # Classify based on compound score thresholds
    if score['compound'] >= 0.05:
        return 'good'
    elif score['compound'] <= -0.05:
        return 'bad'
    else:
        return 'neutral'

# Apply the sentiment analysis function to the 'Review' column
df['Sentiment'] = df['Review'].apply(get_sentiment)

# Display the sentiment distribution
print("Distribution of Sentiments:")
display(df['Sentiment'].value_counts())
```

```python
plt.figure(figsize=(8, 7))
ax = sns.countplot(x='Sentiment', data=df, hue='Sentiment', palette='pastel', order=['good', 'neutral', 'bad'], legend=False)
plt.title('Overall Sentiment Distribution of Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')

# Add annotations
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```
